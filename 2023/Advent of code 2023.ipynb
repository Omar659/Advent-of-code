```python
with open("testo5.txt", "r") as f:
    testo = ""
    for row in f.readlines():
        testo += row

seeds = [int(x) for x in testo.split("\n\n")[0].split(":")[1].strip().split(" ")]
seed_to_soil = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[1].split(":")[1].strip().split("\n")]
soil_to_fertilizer = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[2].split(":")[1].strip().split("\n")]
fertilizer_to_water = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[3].split(":")[1].strip().split("\n")]
water_to_light = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[4].split(":")[1].strip().split("\n")]
light_to_temperature = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[5].split(":")[1].strip().split("\n")]
temperature_to_humidity = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[6].split(":")[1].strip().split("\n")]
humidity_to_location = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[7].split(":")[1].strip().split("\n")]

def get_value_from_map(map, sources):
    all_destinations = []
    for source in sources:
        destinations = []
        for map_to_destination in map:
            destination_map = map_to_destination[0]
            source_map = map_to_destination[1]
            range_map = map_to_destination[2]
            if source_map <= source and source_map + range_map - source >= 0:
                destinations.append(destination_map - source_map + source)
        if destinations == []:
            destinations = [source]
        all_destinations += destinations
        # print()
    return all_destinations

soils = get_value_from_map(seed_to_soil, seeds)
fertilizers = get_value_from_map(soil_to_fertilizer, soils)
waters = get_value_from_map(fertilizer_to_water, fertilizers)
lights = get_value_from_map(water_to_light, waters)
temperatures = get_value_from_map(light_to_temperature, lights)
humidities = get_value_from_map(temperature_to_humidity, temperatures)
locations = get_value_from_map(humidity_to_location, humidities)
min(locations)
```

```python
import copy

def create_map(testo, i_map):
    map = [[int(y) for y in x.split(" ")] for x in testo.split("\n\n")[i_map].split(":")[1].strip().split("\n")]
    map.sort(key=lambda x: x[1])
    new_map = []
    if map[0][1] != 0:
        # new_map.append(["new", 0, 0, map[0][2]])
        new_map.append([0, 0, map[0][2]])
    else:
        new_map.append(map[0])
    for i in range(1, len(map), 1):
        if map[i][1] - map[i-1][2] != map[i-1][1]:
            new_destination_source = map[i-1][2] + map[i-1][1]
            new_map.append([new_destination_source, new_destination_source, map[i][1] - new_destination_source])
        new_map.append(map[i])
    new_map.append([map[-1][2] + map[-1][1], map[-1][2] + map[-1][1], 99999999999999999999999999])
    return new_map

def read_problem5():
    with open("testo5.txt", "r") as f:
        testo = ""
        for row in f.readlines():
            testo += row

    tmp_seeds = [int(x) for x in testo.split("\n\n")[0].split(":")[1].strip().split(" ")]
    seeds = []
    for i in range(0, len(tmp_seeds), 2):
        seeds.append([tmp_seeds[i], tmp_seeds[i] + tmp_seeds[i+1] - 1])

    seed_to_soil = create_map(testo, 1)
    soil_to_fertilizer = create_map(testo, 2)
    fertilizer_to_water = create_map(testo, 3)
    water_to_light = create_map(testo, 4)
    light_to_temperature = create_map(testo, 5)
    temperature_to_humidity = create_map(testo, 6)
    humidity_to_location = create_map(testo, 7)


    maps_link = {
#                    MAP                       FROM                        TO
           "humidity_to_location"   :   ["temperature_to_humidity"      , "-"                       ],
        "temperature_to_humidity"   :   ["light_to_temperature"         , "humidity_to_location"    ],
           "light_to_temperature"   :   ["water_to_light"               , "temperature_to_humidity" ],
                 "water_to_light"   :   ["fertilizer_to_water"          , "light_to_temperature"    ],
            "fertilizer_to_water"   :   ["soil_to_fertilizer"           , "water_to_light"          ],
             "soil_to_fertilizer"   :   ["seed_to_soil"                 , "fertilizer_to_water"     ],
                   "seed_to_soil"   :   ["seeds"                        , "soil_to_fertilizer"      ],
                          "seeds"   :   ["-"                            , "seed_to_soil"            ]
    }
    maps = {
        "humidity_to_location":  humidity_to_location,
        "temperature_to_humidity":  temperature_to_humidity,
        "light_to_temperature":  light_to_temperature,
        "water_to_light":  water_to_light,
        "fertilizer_to_water":  fertilizer_to_water,
        "soil_to_fertilizer":  soil_to_fertilizer,
        "seed_to_soil":  seed_to_soil,
        "seeds": seeds
    }
    return maps, maps_link

class Node():
    def __init__(self, value, tab, name):
        self.value = value
        self.tab = tab
        self.name = name
        self.childs = []

    def __copy__(self):
        new_node = Node(
            copy.deepcopy(self.value),
            copy.deepcopy(self.tab),
            copy.deepcopy(self.name),
        )
        new_node.childs = copy.deepcopy(self.childs)
        return new_node

    def __str__(self):
        s = self.name + ": " + str(self.value) + "\n"
        for child in self.childs:
            s += "\t"*self.tab + str(child) + "\n"
        return s[:-1]

    def __repr__(self):
        s = str(self.value) + "\n"
        for child in self.childs:
            s += "\t"*self.tab + self.name + ": " + str(child) + "\n"
        return s[:-1]

class Tree():
    def __init__(self, maps, maps_link):
        self.maps = maps
        self.maps_link = maps_link

    def intersection(self, moving, fixed):
        # Non si intersecano
        if moving[1] < fixed[0] or moving[0] > fixed[1]:
            return [], None
        # Si intersecano
        # Inizio intersezione
        if moving[0] < fixed[0]:
            start = fixed[0]
            offset_start = 0
        else:
            start = moving[0]
            offset_start = moving[0] - fixed[0]
        # Fine intersezione
        if moving[1] > fixed[1]:
            end = fixed[1]
            offset_end = 0
        else:
            end = moving[1]
            offset_end = fixed[1] - moving[1]
        return [start, end], [offset_start, offset_end]

    def build_graph(self, from_, to_, current, tab = 1):
        self.root = self.__build_graph(from_, to_, current, tab)

    def __build_graph(self, from_, to_, current, tab):
        root = Node([from_, to_], tab, current)
        next = self.maps_link[current][0]
        if self.maps.get(next) is None:
            return root
        childs = []
        if next == "seeds":
            for destination_map in self.maps[next]:
                intersection, offsets = self.intersection(destination_map, from_)
                if intersection == []:
                    continue
                childs.append([destination_map, destination_map])
        else:
            for map_to_destination in self.maps[next]:
                range_map = map_to_destination[2] - 1
                destination_map = [map_to_destination[0], map_to_destination[0] + range_map]
                source_map = [map_to_destination[1], map_to_destination[1] + range_map]

                intersection, offsets = self.intersection(destination_map, from_)
                if intersection == []:
                    continue
                childs.append([destination_map, source_map])
        for child in childs:
            child = self.__build_graph(child[1], child[0], next, tab+1)
            root.childs.append(child)
        return root

    def all_paths(self):
        stack = [(self.root, [self.root])]
        self.paths = []
        while stack:
            current_node, current_path = stack.pop()
            if current_node.name == "seeds":
                self.paths.append(current_path)
            else:
                for child in current_node.childs:
                    stack.append((child, current_path + [child]))
        for i in range(len(self.paths)):
            self.paths[i].reverse()

    def print_path(self, path):
        for node in path:
            print(node.name, node.value)

    def check_paths(self):
        locations = []
        for path_i, path in enumerate(self.paths):
            ok = True
            new_path = copy.deepcopy(path)
            for node_idx in range(len(new_path)-1):
                new_source, offsets = self.intersection(new_path[node_idx].value[1], new_path[node_idx+1].value[0])
                if new_source == []:
                    new_value = [[-1, -1], [-1, -1]]
                    ok = False
                else:
                    new_destination = [new_path[node_idx+1].value[1][0] + offsets[0], new_path[node_idx+1].value[1][1] - offsets[1]]
                    new_value = [new_source, new_destination]
                new_path[node_idx+1].value = new_value
            self.paths[path_i] = new_path
            if ok:
                locations.append(new_path[-1].value[1][0])
        return min(locations) if locations != [] else 1000000000000000000




maps, maps_link = read_problem5()
g = Tree(maps, maps_link)

min_ = 1000000000000000000
for i, humidity_map in enumerate(maps["humidity_to_location"]):
    print("\r" + str(i) + "/" + str(len(maps["humidity_to_location"]) - 1), end="")
    range_map = humidity_map[2] - 1
    destination_map = [humidity_map[0], humidity_map[0] + range_map]
    source_map = [humidity_map[1], humidity_map[1] + range_map]
    g.build_graph(source_map, destination_map, "humidity_to_location")
    g.all_paths()
    loc = g.check_paths()
    if loc < min_:
        min_ = loc
    # print(min_)
    # break
min_
```

```python
def ricerca_dicotomica(start, end, time, travel):
    while True:
        speed_centro = (start + end) // 2
        if (time - speed_centro)*speed_centro > travel:
            end = speed_centro - 1
        elif (time - speed_centro)*speed_centro < travel:
            start = speed_centro + 1
        if start >= end:
            return speed_centro

race = [48989083, 390110311121360]
boundary = race[0]//2 + 1
minmil = ricerca_dicotomica(1, boundary, race[0], race[1])
f = lambda s: (race[0] - s)*s
minmil = 10007574
maxmil = race[0] - minmil
f(minmil) >= race[1], f(maxmil) >= race[1], minmil, maxmil, maxmil-minmil+1
```
